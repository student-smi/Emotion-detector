```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as mtp 
import seaborn as sb
```

```python
df = pd.read_csv('train.txt' , sep=';' , header=None , names=['text' , 'emotion'])
```

```python
df.head()
```

```python
df_emo_num=df['emotion'].unique()
```

```python
df_emo_num
```

```python
i = 0
unique_emotions_num = {}
for emo  in  df_emo_num: 
    unique_emotions_num[emo] = i 
    i += 1


unique_emotions_num

    
    
```

```python
df['emotion'] = df['emotion'].map(unique_emotions_num)
```

```python
df.value_counts()
```

```python


df['text'] = df['text'].apply( lambda x:x.lower())
```

```python
df.head()
```

```python
import nltk
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python

nltk.download('punkt')
nltk.download('stopwords')
```

```python
stopwords = set(stopwords.words('english')) 
```

```python
stopwords
```

```python


def remove(txt):
    clean = [] 
    words = txt.split()

    for i in words: 
        if not i in stopwords:
     
            clean.append(i)

    return ' '.join(clean)
 
```

```python
df['text'] = df['text'].apply(remove)
```

```python
df.head()
```

```python
df.loc[0]['text']
```

```python
from sklearn.model_selection import train_test_split
```

```python
 X_train, X_test, y_train, y_test = train_test_split(df['text'], df['emotion'], test_size=0.20, random_state=42)
```

```python
X_train
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer , CountVectorizer
```

```python
bow_vector = CountVectorizer()
```

```python
X_Train_Bag = bow_vector.fit_transform(X_train)
```

```python
X_test_bag = bow_vector.transform( X_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
```

```python
model = MultinomialNB()
```

```python
model.fit(X_Train_Bag , y_train)
```

```python
y_pre = model.predict(X_test_bag)
```

```python
acc = accuracy_score(y_test , y_pre)
```

```python
acc
```

```python
tf_idf_vector = TfidfVectorizer()
```

```python
X_Train_tf_idf = tf_idf_vector.fit_transform(X_train)
X_Test_tf_idf = tf_idf_vector.transform(X_test)


```

```python
nb2_model = MultinomialNB()
```

```python
nb2_model.fit(X_Train_tf_idf , y_train)
```

```python
Y_Pre_tf_idf = nb2_model.predict(X_Test_tf_idf)
```

```python
result = accuracy_score(y_test , Y_Pre_tf_idf)
result
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model_lr = LogisticRegression()
```

```python
model_lr.fit(X_Train_tf_idf , y_train)
```

```python
y_pre_lr = model_lr.predict(X_Test_tf_idf)
```

```python
result_lr = accuracy_score(y_test , y_pre_lr)
result_lr
```
